```python
import csv

with open('ds_salaries.csv', mode='r') as file:
    csv_reader = csv.reader(file)
    for row in csv_reader:
        print(row)
```

```python
import numpy as np
data = np.genfromtxt('ds_salaries.csv', delimiter=',', skip_header=1)
```

```python
import csv

with open('ds_salaries.csv', mode='r') as file:
    csv_reader = csv.DictReader(file)
    for row in csv_reader:
        print(row)
```
